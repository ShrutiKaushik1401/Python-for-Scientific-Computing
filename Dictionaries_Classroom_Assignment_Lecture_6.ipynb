```python
#Dictionaries
# 1 - Import	paylab
import numpy as np
import pylab as plb
import matplotlib.pyplot as plt


#2 - Create	a	dictionary	with	5	keys	and	empty	values	in	A

A = {'Car': '','Bus': '','Plane': '','Ship': '', 'Bike':''}
print('Empty Dict A is:', A)

#3 - Using	a	function,	assign	random	values	to	each	key	between	[0:10],
#using	a	for	loop	and	return	the	result	in	B
def assign_dict_val():
    for key,value in A.items():
        value = np.random.randint(0,10)
        A[key] = value             
    return A

B = assign_dict_val()
print('Random Value Dict B is:', B)



#4 - using a function, change the value of any member in B that is less than 5 with the result from 4.1
#(consider using an if statement in a loop):
    #4.1 - using normal distribution with mean = 2 and std = 3 create an array of size 256 points
    #4.2 - using histogram with 12 bins, plot the result from 4.1 with a pause of 1 sec. use proper
    #labeling(figure,title,labels,legend,grid etc.)
def change_dict_val(B): 
    normal_dist = plb.normal(2,3,size=256) 
    plb.figure(1)
    plb.hist(normal_dist,normed=True,bins=12,histtype="stepfilled",color="green",label="Uniform",alpha=0.65) 
    plb.title("Normal Distribution/ Histogram")
    plb.legend(loc="upper left")    
    plb.xlabel("Value")
    plb.ylabel("Frequency")
    plb.grid()
    plb.pause(1)
    
    for key,value in B.items():
        if B[key] < 5:
            B[key] = normal_dist
    return B

# 5 - Assign	the	result	from	4.	in	C
C = change_dict_val(B)
print('Normal Dist Dict C is:', C) 


#6 -  Update	one	of	the	keys	in	C	with	another	using	the	pop	feature
A["Bike"] = A.pop("Car")
print('New popped dict C is:',A)

#7 -  Update	another	key	in	C	manually	(add	the	new	one	and	delete	the	old	one)	
A["Car"]= A["Bus"]
del A["Bus"]
print('Updated dict C is:',A)

#8 - Compare	A,	B	and	C	using	a	short	conditional	expression
print("Check to see if A, B and C are equal:", A is B is C)

print('Shruti Kaushik is done with Lecture 6 Classroom Assignment')
```
